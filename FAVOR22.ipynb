```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
pd.read_csv("bmi_train.csv")
```

```python
#Reading data
data=pd.read_csv("bmi_train.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
#checking if there is missing data
data.isnull().sum()
```

```python
data.shape
```

```python
# Scatter plot
data.plot(kind='scatter', x="Height", y="Weight", s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#checking if there is missing data
data.isnull().sum()
```

```python
x=data[["Height"]]
y=data[["Weight"]]
```

```python
model=linear_model.LinearRegression()
```

```python
#splitting the data into 70% training and 30% testing
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
#training the model

model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test, y_test, label='Actual Values', color='blue', marker='o')

# Plot predicted values
plt.plot(x_test, y_pred, label='Predicted Values', color='red', marker='x')
```

```python
model.intercept_
```

```python
model.coef_
```

```python
predicted_weigtht= model.predict([[8]])
print(predicted_weigtht)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("Mean squared error")
print(mean_squared_error(y_test,y_pred))
print("R_squared")
print(r2_score(y_test,y_pred))
print("Mean absolute error")
print(mean_absolute_error(y_test,y_pred))
```
